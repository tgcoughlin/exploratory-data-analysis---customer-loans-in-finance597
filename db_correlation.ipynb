```python
import pandas as pd

new_df = pd.read_hdf('latest_df_with_dtypes.h5', key='data')

```

```python
#Run a correlation matrix
numeric_columns = new_df.select_dtypes(include=['number'])

correlation_matrix = numeric_columns.corr()

print(correlation_matrix)
```

```python
#HeatMap
import plotly.express as px
numeric_columns = new_df.select_dtypes(include=['number'])
px.imshow(numeric_columns.corr(), title="Correlation heatmap of student dataframe")
```

```python
columns_to_drop = ['funded_amount', 'funded_amount_inv', 'instalment']
df_colin_dropped = new_df.drop(columns=columns_to_drop)

df_colin_dropped.info()
```

```python
object_columns = df_colin_dropped.select_dtypes(include=['object']).columns
df_colin_dropped[object_columns] = df_colin_dropped[object_columns].astype(str)

df_colin_dropped.to_hdf('colinea_dropped_df.h5', key='data', mode='w', format='table')
```
